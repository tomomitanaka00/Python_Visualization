```python
movies=pd.read_csv("P4-Movie-Ratings.csv")
```

```python
movies.columns=["Film", "Genre", "CriticRating", "AudienceRating", "BudgetMillions", "Year"]
```

```python
movies.Film=movies.Film.astype('category')
movies.Genre=movies.Genre.astype('category')
movies.Year=movies.Year.astype('category')
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
vis1=sns.lmplot(data=movies,x="CriticRating", y="AudienceRating", fit_reg=False,hue="Genre",size=7,aspect=1)
```

```python
#karnel density estimates
kl=sns.kdeplot(movies.CriticRating, movies.AudienceRating, shade=True,shade_lowest=False,  cmap='Reds')
```
